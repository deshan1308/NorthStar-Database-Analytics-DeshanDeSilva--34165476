```python
pipeline = [
    {
        "$group": {
            "_id": "$hub.zone",
            "avg_distance": {"$avg": "$route.distance_km"},
            "delivery_count": {"$sum": 1}
        }
    },
    {
        "$sort": {"avg_distance": -1}
    }
]

agg_result = list(deliveries_col.aggregate(pipeline))
for doc in agg_result:
    print(doc)
```
